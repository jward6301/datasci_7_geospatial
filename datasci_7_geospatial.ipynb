# Geocoding

Dataframe 1
* This dataframe provides informaiton on birth rates in Lake County, Illinois. The columns idenftiy percentages of low birth weights, preterm births, teen births, number of live births (birth rate), and how many appointments and care provided during the first 13 weeks of pregnancy (F1stTricare).

```python
df1 = gpd.read_file('Birth_Statistics.geojson')
df1
```

```python
df1.columns
```

```python
missing_values = df1.isnull()
print(missing_values)
```

```python
df1.plot("Preterm", legend=True)
```

About Dataframe 1: I analyzed based off of preterm births. The data seems to be scattered throughout the area of interest with almsot every area being in the middle percentage.

Dataframe 2
* This dataset provides information regarding health insurance coverage rates in the Detroit Tri-county region. The rates were taken in 2014 by zip code provided.

```python
df2 = gpd.read_file('HealthInsuranceCoverage.geojson')
df2
```

```python
df2.columns
```

```python
missing_values = df2.isnull()
print(missing_values)
```

```python
df2.dropna(inplace=True)
df2
```

```python
df2.plot("NoHealthInsurance", legend=True)
```

About Dataframe 2: This was analyzed based on rates of no health insurance. It seems that the higher lattitude areas and some middle areas have lower rates of no health insurance. The lower lattitude regions have higher rates of no health insurance.
